```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Visualize Merged DataFrame
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Assign Values to Variables
total_schools = len(school_data_complete["school_name"].unique()) # verified
total_students = len(school_data_complete["student_name"]) # verified
total_budget = school_data_complete["budget"].unique().sum() #verified
avg_math_score = school_data_complete["math_score"].mean() #verified
avg_reading_score = school_data_complete["reading_score"].mean() #verified
overall_passing_score = (avg_math_score + avg_reading_score) / 2 #verified
num_std_math_passing = len(school_data_complete.loc[school_data_complete["math_score"] >= 70]) #verified
pct_math_passing = (num_std_math_passing / total_students) * 100 #verified
num_std_reading_passing = len(school_data_complete.loc[school_data_complete["reading_score"] >= 70]) #verified
pct_reading_passing = (num_std_reading_passing / total_students) * 100 #verified

# Create Summary Data
PyCity_Summary_data = {'Total Schools': [total_schools], 'Total Students': [total_students], 
                      'Total Budget': [total_budget],'Average Math Score': [avg_math_score],
                      'Average Reading Score': [avg_reading_score], '% Passing Math': [pct_math_passing],
                      '% Passing Reading': [pct_reading_passing], '% Overall Passing Rate': [overall_passing_score]} 
  
# Create DataFrame 
PyCity_Summary_df = pd.DataFrame(PyCity_Summary_data) 
  
# Format the Total Students and Total Budget Values
PyCity_Summary_df["Total Students"] = PyCity_Summary_df["Total Students"].map("{:,}".format)
PyCity_Summary_df["Total Budget"] = PyCity_Summary_df["Total Budget"].map("${:,.2f}".format)

# Print the DataFrame
PyCity_Summary_df

```
